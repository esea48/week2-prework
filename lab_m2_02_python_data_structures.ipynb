## Warm-Up: Data Types and Operators

### Exercise 1: Data Types

#### 1.1 Check the data type of each variable using `type()`

```python
x1 = 1.1
x2 = "Ironhack"
x3 = "1.1"
x4 = True
x5 = "True"
x6 = -1

print("x1:", type(x1))  # float
print("x2:", type(x2))  # str
print("x3:", type(x3))  # str
print("x4:", type(x4))  # bool
print("x5:", type(x5))  # str
print("x6:", type(x6))  # int
```

#### 1.2 Difference between `x1` and `x3`

`x1 = 1.1` is a **float** — a numeric value Python can do math with.  
`x3 = "1.1"` is a **string** — just text. It looks like a number but Python treats it as characters, so you can't do arithmetic with it directly.

#### 1.3 What happens if you subtract `x3` from `x1`?

```python
# Uncommenting the line below shows what happens:
# x1 - x3

# This raises a TypeError: unsupported operand type(s) for -: 'float' and 'str'
# Python cannot subtract a string from a float. They are different types.
print("x1 - x3 raises: TypeError — can't subtract a string from a float.")
```

#### 1.4 Difference between `x4` and `x5`

`x4 = True` is a **bool** — a Python boolean value (True or False).  
`x5 = "True"` is a **string** — the word "True" written as text. Python does not treat it as a logical value.

#### 1.5 What happens if you subtract `x5` from `x4`?

```python
# Uncommenting the line below shows what happens:
# x4 - x5

# This raises a TypeError: unsupported operand type(s) for -: 'bool' and 'str'
# Even though True has a numeric value (1) internally, Python won't subtract a string from it.
print("x4 - x5 raises: TypeError — can't subtract a string from a bool.")
```

## Exercise 1: Working with Lists

**Scenario:** Creating and managing a guest list for the event.  
**Why a list?** Like seats in a theater — ordered, numbered, and the same person can appear more than once (e.g., if they RSVP twice).

```python
# 1. Create a list with 5 guest names
guest_list = ["Alice", "Bob", "Charlie", "Diana", "Eve"]
print("Initial guest list:", guest_list)
```

```python
# 2. Add two more guests using .append()
guest_list.append("Frank")
guest_list.append("Grace")
print("After adding two guests:", guest_list)
```

```python
# 3. Print total number of guests
print("Total guests:", len(guest_list))
```

```python
# 4. Print the first and last guest using indexing
# Index 0 = first item; index -1 = last item (counting from the end)
print("First guest:", guest_list[0])
print("Last guest:", guest_list[-1])
```

```python
# 5 & 6. Sort alphabetically and print
guest_list.sort()
print("Sorted guest list:", guest_list)
```

**Key takeaway:** Lists maintain insertion order and allow duplicates. You access items by their position (index), starting at 0.

## Exercise 2: Working with Dictionaries

**Scenario:** Storing contact info for each guest.  
**Why a dictionary?** Like a phone book — you look up a person by name (key) to find their number (value), not by flipping to page 3.

```python
# 1. Create a dictionary with 3 key-value pairs (name -> email)
contact_info = {
    "Alice": "alice@email.com",
    "Bob": "bob@email.com",
    "Charlie": "charlie@email.com"
}
print("Initial contacts:", contact_info)
```

```python
# 2. Add 2 more guests and their emails
contact_info["Diana"] = "diana@email.com"
contact_info["Eve"] = "eve@email.com"
print("After adding two contacts:", contact_info)
```

```python
# 3. Look up a specific guest's email by name
print("Alice's email:", contact_info["Alice"])
```

```python
# 4. Print all keys (guest names)
print("All guest names:", list(contact_info.keys()))
```

```python
# 5. Print all values (email addresses)
print("All emails:", list(contact_info.values()))
```

```python
# 6. Check if a specific guest exists using the 'in' keyword
print("Is 'Alice' in contacts?", "Alice" in contact_info)   # True
print("Is 'Zara' in contacts?", "Zara" in contact_info)     # False
```

**Key takeaway:** Dictionaries pair keys with values for fast lookups. Keys must be unique — adding the same key again just overwrites the old value.

## Exercise 3: Working with Sets

**Scenario:** Collecting unique guest interests to plan activities.  
**Why a set?** Like a bag of raffle tickets — you only keep one ticket per person, duplicates are automatically discarded, and there's no fixed order.

```python
# 1. Create a set with 5 interests
interests = {"music", "reading", "sports", "cooking", "travel"}
print("Initial interests:", interests)
```

```python
# 2. Add 2 more interests using .add()
interests.add("photography")
interests.add("gaming")
print("After adding two interests:", interests)
```

```python
# 3. Try adding a duplicate — sets silently ignore it
interests.add("music")  # 'music' is already in the set
print("After adding duplicate 'music':", interests)
print("Size unchanged — sets only store unique values.")
```

```python
# 4. Notice the set doesn't maintain insertion order
print("Set (no guaranteed order):", interests)
```

```python
# 5. Create a second set with some overlapping interests
other_interests = {"cooking", "travel", "hiking", "painting", "music"}
```

```python
# 6 & 7. Find common interests using .intersection() and the & operator
common = interests.intersection(other_interests)
common_alt = interests & other_interests  # same result, shorter syntax

print("Common interests (using .intersection()):", common)
print("Common interests (using &):", common_alt)
```

**Key takeaway:** Sets store only unique elements and have no guaranteed order. They're ideal when you need to remove duplicates or find what two groups have in common.

## Comparison Reflection

**1. When would you use a list?**  
Use a list when order matters and/or duplicates are allowed. For example, a queue of arriving guests, a playlist of songs, or steps in a recipe — things where position is meaningful and the same item might appear more than once.

**2. When would you use a dictionary?**  
Use a dictionary when you need to look something up by a meaningful label (key) rather than by position. For example, mapping student names to grades, product IDs to prices, or guest names to emails. Think: "I know *what* I'm looking for, not *where* it is."

**3. When would you use a set?**  
Use a set when uniqueness is what matters — removing duplicates from a collection, checking membership quickly, or comparing two groups to find overlaps or differences. For example, unique visitors to a website, or common tags between two blog posts.

**4. What's the main difference between a list and a set?**  
A list is ordered and allows duplicates; a set is unordered and only stores unique elements. Think of a list as a numbered queue (position matters, same person can be in it twice) and a set as a stamp collection (you either have a stamp or you don't, and the drawer order doesn't matter).

**5. Can you access dictionary values by index like you do with lists? Why or why not?**  
No. Dictionaries are accessed by *key*, not by position. There is no "item at index 2" in a dictionary — you ask for `contact_info["Alice"]`, not `contact_info[0]`. This is because dictionaries are designed for named lookups, not sequential access. (Since Python 3.7 dicts do preserve insertion order, but the design pattern is still key-based access.)
