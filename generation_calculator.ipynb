# Exercise 1 - Generation Calculator

Create a function named `generation_calculator` that returns the generation a person belongs to based on their birth year.

| Birth Year | Generation |
|---|---|
| < 1949 | Silent Generation |
| 1949–1968 | Baby Boomer |
| 1969–1980 | Generation X |
| 1981–1993 | Millennial |
| >= 1994 | Generation Z |

```python
def generation_calculator(birth_year):
    """
    Determines the generation a person belongs to based on their birth year.
    
    Parameters:
        birth_year (int): The year the person was born.
    
    Returns:
        str: The name of the generation.
    """
    if birth_year < 1949:
        return "Silent Generation"
    elif birth_year <= 1968:
        return "Baby Boomer"
    elif birth_year <= 1980:
        return "Generation X"
    elif birth_year <= 1993:
        return "Millennial"
    else:
        return "Generation Z"
```

```python
# Test the function with different birth years
test_years = [1940, 1955, 1975, 1988, 2000]

for year in test_years:
    result = generation_calculator(year)
    print(f"Birth year {year}: {result}")
```
