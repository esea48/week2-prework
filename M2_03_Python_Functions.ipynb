## Exercise 1: Greeting Function (From Scratch)

Create a function called `greet_person` that:
- Takes a `name` parameter and an optional `language` parameter (default: `"Python"`)
- Returns a personalized greeting
- Mentions the language only if it differs from the default

````python
def greet_person(name, language="Python"):
    """
    Greets a person by name and optionally mentions their favourite language.

    Parameters:
        name (str): The person's name.
        language (str): Their favourite programming language. Defaults to 'Python'.

    Returns:
        str: A personalised greeting message.
    """
    # Start with the base greeting
    message = f"Hello, {name}! Welcome to Python programming!"

    # Only mention the language if it's different from the default
    if language != "Python":
        message += f" I see you're interested in {language} too!"

    return message
````

````python
# --- Test cases ---

# Test 1: No language provided → uses default 'Python'
print(greet_person("Alice"))

# Test 2: Different language provided → should mention JavaScript
print(greet_person("Bob", "JavaScript"))

# Test 3: Language is explicitly 'Python' → same as default, no extra line
print(greet_person("Charlie", "Python"))
````

## Exercise 2: Grade Calculator (From Pseudocode)

Pseudocode provided:
```
FUNCTION calculate_grade(percentage):
    IF percentage >= 90  → RETURN "A"
    ELSE IF percentage >= 80  → RETURN "B"
    ELSE IF percentage >= 70  → RETURN "C"
    ELSE IF percentage >= 60  → RETURN "D"
    ELSE  → RETURN "F"
END FUNCTION
```

We check from **highest to lowest** so each condition only triggers when all higher ones have already failed — like a waterfall that stops at the first ledge it hits.

````python
def calculate_grade(percentage):
    """
    Converts a percentage score to a letter grade.

    Parameters:
        percentage (int or float): Score between 0 and 100.

    Returns:
        str: Letter grade (A, B, C, D, or F), or an error message for invalid input.
    """
    # Bonus: validate input before doing anything else
    if percentage < 0 or percentage > 100:
        return "Invalid percentage. Please enter a value between 0 and 100."

    # Check grades from highest to lowest (order matters!)
    if percentage >= 90:
        return "A"
    elif percentage >= 80:
        return "B"
    elif percentage >= 70:
        return "C"
    elif percentage >= 60:
        return "D"
    else:
        return "F"
````

````python
# --- Test cases ---

print(calculate_grade(95))   # → A
print(calculate_grade(85))   # → B
print(calculate_grade(75))   # → C
print(calculate_grade(65))   # → D
print(calculate_grade(55))   # → F

# Edge cases
print(calculate_grade(90))   # → A  (exactly on boundary)
print(calculate_grade(60))   # → D  (exactly on boundary)

# Bonus: invalid inputs
print(calculate_grade(-5))   # → error message
print(calculate_grade(105))  # → error message
````

## Reflection Questions

**1. What is the main advantage of using functions instead of writing the same code multiple times?**

Functions let you write logic once and reuse it everywhere — like a reusable stamp instead of hand-writing the same address repeatedly. If you need to change the logic, you change it in one place and every caller automatically gets the update.

**2. In Exercise 1, why did we use a default parameter value for `language`?**

A default value makes the parameter *optional*. Callers who don't care about it don't have to pass anything, which keeps the function easy to use in the common case while still supporting the full case when needed.

**3. In Exercise 2, why is it important to check conditions from highest to lowest (90, then 80, then 70...)?**

Python evaluates `if/elif` chains top-to-bottom and stops at the first `True`. If we started with `>= 60`, a score of 95 would match immediately and return `"D"` — wrong! Starting high means each condition only fires once all stricter thresholds have already been ruled out.

**4. How did the pseudocode help you in Exercise 2? What would have been different without it?**

Pseudocode acted like architectural blueprints — it settled *what* the logic should do before we had to think about *how* Python expresses it. Without it, we'd have to design the structure and write the syntax at the same time, which is like trying to design and build a house simultaneously. The pseudocode let us focus each step separately, reducing errors.
